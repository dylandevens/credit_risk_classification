```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import confusion_matrix, classification_report
```

---

## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
file_path = Path("Resources/lending_data.csv")
df_all_data = pd.read_csv(file_path)

# Review the DataFrame
print(df_all_data.info())
print()
print(f"shape =  {df_all_data.shape}")
print()
print(f"columns = {list(df_all_data.columns)}")
```

```python
df_all_data.head(5)
```

### Step 2: Create the labels set (`y`)  from the “loan_status” column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features
y = df_all_data['loan_status']

# Separate the y variable, the labels
df_data = df_all_data.copy()
df_data = df_data.drop(columns='loan_status', axis=1)

# Separate the X variable, the features
X = df_data
```

```python
# Review the y variable Series
print(len(y))
print(y[:5])
```

```python
# Review the X variable DataFrame
print(X.shape)
X.head(3)
```

### Step 3: Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

---

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
classifier = LogisticRegression(random_state=1, solver='lbfgs', max_iter=100, verbose=1)
classifier
```

```python
# Fit the model using training data
classifier.fit(X_train, y_train)
```

```python
# Model Score
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python
# Make a prediction using the testing data
predictions = classifier.predict(X_test)
df_predictions = pd.DataFrame({"Prediction": predictions, "Actual": y_test})
df_predictions
```

### Step 3: Evaluate the model’s performance by doing the following:

* Generate a confusion matrix.

* Print the classification report.

```python
# Generate a confusion matrix for the model
c_m = confusion_matrix(y_test, predictions)
c_m
# tn, fp, fn, tp
```

```python
df_cm = pd.DataFrame(c_m, index=["test_negative", "test_positive"], columns=["pred_negative", "pred_positive"])
df_cm
```

```python
# Print the classification report for the model
target_names = ["healthy loan (0)", "high-risk loan (1)"]
print(classification_report(y_test, predictions, target_names=target_names))
```

```python
# import dependecies 
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import  accuracy_score
```

```python
# Separate the data into labels and features
y2 = df_all_data['loan_status']

# Separate the y variable, the labels
df_data2 = df_all_data.copy()
df_data2 = df_data2.drop(columns='loan_status', axis=1)

# Separate the X variable, the features
X2 = df_data2
```

```python
# Review the y variable Series
print(len(y2))
print(y2[:5])
```

```python
# Review the X variable DataFrame
print(X2.shape)
X2.head(3)
```

```python
# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, random_state=1)
```

### SCALE

```python
# Creating StandardScaler instance
scaler = StandardScaler()
# Fitting Standard Scaller
X_scaler = scaler.fit(X_train2)
# Scaling data
X_train_scaled = X_scaler.transform(X_train2)
X_test_scaled = X_scaler.transform(X_test2)


```

### Model creation, fit, predict

```python
# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
classifier2 = LogisticRegression(random_state=1, solver='lbfgs', max_iter=100, verbose=1)

# Fit the model using training data
classifier2.fit(X_train_scaled, y_train2)

# Making predictions using the testing data
predictions2 = classifier2.predict(X_test_scaled)

```

### Scores and results

```python
# Model Score
print(f"Training Data Score: {classifier2.score(X_train_scaled, y_train2)}")
print(f"Testing Data Score: {classifier2.score(X_test_scaled, y_test2)}")

# Generate a confusion matrix for the model
c_m2 = confusion_matrix(y_test2, predictions2) # tn, fp, fn, tp
print()
print("confusion matrix for scaled data is:")
print(c_m2)
print()

df_cm2 = pd.DataFrame(c_m2, index=["test_negative", "test_positive"], columns=["pred_negative", "pred_positive"])
df_cm2
```

```python
# Print the classification report for the model
target_names = ["healthy loan (0)", "high-risk loan (1)"]
print(classification_report(y_test2, predictions2, target_names=target_names))
```

Based on the precision scores, 100% of predicted healthy loans(0) were correct, and 84% of high-risk loans(1) were predicted correctly.

Based on the recall scores, out of all the healthy loans(0) in the test dataset, 99% of them were correctly predicted, and out of all the high-risk-loans(1), 99% of them were correclty predicted.
